### Ditricts data

```python
#selecting data for districts for each year

district_data_2011 =  data_2011[list(data_2011.columns)[1:]]

district_data_2012 =  data_2012[list(data_2012.columns)[1:]]

district_data_2013 =  data_2013[list(data_2013.columns)[1:]]

district_data_2014 =  data_2014[list(data_2014.columns)[1:]]

district_data_2015 =  data_2015[list(data_2015.columns)[1:]]
```

```python
year="2011"

# Selecting data and grouping by district
district_data_2011.groupby("District_Name").sum().sort_values(by=list(district_data_2011.columns)[3:],ascending=False).head(4).plot(kind="bar",figsize=[30,10],subplots=False,fontsize=17)
plt.title("A GRAPH SHOWING PERFORMANCE OF THE BEST FOUR DISTRICTS  IN 2011")
print()
```

```python
year="2015"
district_data_2015.groupby("District_Name").sum().sort_values(by=list(district_data_2015.columns)[3:],ascending=False).head(4).plot(kind="bar",figsize=[30,10],subplots=False,fontsize=17)
plt.title("A GRAPH SHOWING PERFORMANCE IN 2015")
print()
```

### Selecting trend in performance in Studentsperformance in after change of curriculum 

```python
#doing summation of students performance above 16 points 
performance_2013 = int(data_2013[["%16-20 Points.2"]].sum())
performance_2014 = int(data_2014[["%16-20 Points.3"]].sum())
performance_2015 = int(data_2015[["%16-20 Points.4"]].sum())
```

```python
performance_2015
```

```python
#selecting important data of each year
x = [2013, 2014,2015] 

#values to be plotted
y =[performance_2013,performance_2014,performance_2015]
plt.plot(x,y,marker="o")
plt.xlabel("YEARS")
plt.ylabel("%0-25 Points")
plt.title("A LINE GRAPH SHOWING TREND IN PERFORMANCE SINCE CHANGE IN GRADING")
plt.xticks(x,["2013","2014","2015"])
plt.show()
```
